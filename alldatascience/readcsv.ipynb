```python
###pandas concatenate
import pandas as pd
india_weather=pd.DataFrame({
    'City':['mumbai','babcok','banglore'],
    'Temperature':[17,16,30],
    'Humidity':[14,15,17]   
})
india_weather
```

```python
Us_weather=pd.DataFrame({
    'City':['Chicago','Newyork','Atlanta'],
    'Temperature':[18,13,32],
    'Humidity':[11,18,19]   
})
Us_weather
```

```python
df=pd.concat([india_weather,Us_weather],ignore_index=True)
df
```

```python
df=pd.concat([india_weather,Us_weather],keys=['india','us'])
df
```
