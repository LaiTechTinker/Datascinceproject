```python
from sklearn import svm,datasets
iris=datasets.load_iris()
```

```python
import pandas as pd
df =pd.DataFrame(iris.data,columns=iris.feature_names)
df['flowers']=iris.target
df['flowers']=df['flowers'].apply(lambda x:iris.target_names[x])
df
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
X_train,X_test,Y_train,Y_test=train_test_split(iris.data,iris.target, test_size=0.3)
model=svm.SVC(kernel='rbf',C=30,gamma='auto')
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(svm.SVC(kernel='linear',C=10,gamma='auto'),iris.data,iris.target,cv=5)
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=10,gamma='auto'),iris.data,iris.target,cv=5)
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=20,gamma='auto'),iris.data,iris.target,cv=5)
```

```python
kernel=['rbf','linear']
C=[10,20]
avg_score={}
for ker in kernel:
    for cval in C:
        cv_scores=cross_val_score(svm.SVC(kernel=ker,C=cval,gamma='auto'),iris.data,iris.target,cv=5)
        avg_score[ker + '-' + str(cval)]=np.average(cv_scores)
avg_score    
```

```python
# hyper prameter using gridSearchCV
from sklearn.model_selection import GridSearchCV
clf=GridSearchCV(svm.SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},cv=5, return_train_score=False)
clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
df=pd.DataFrame(clf.cv_results_)
df
```
